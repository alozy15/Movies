### The following functions will be used in our exploratory analysis

```python
def dir_film_avg(column1, column2, column3):
    return(groupby_sum(column1, column2) / val_counts(column3)) 

# returns the average revenue per film int / obj
```

```python
#min and max functions
    #taking the index value of the highest and lowest nnumber for searched column

def min(column_name):   
          
    min_id = df[column_name].idxmin()
    min_stats = pd.DataFrame(df.loc[min_id])
           
    return min_stats 
```

```python
def max(column_name):
    
    
    max_id = df[column_name].idxmax()
    #calling by index number,storing that row info to a variable
    max_stats = pd.DataFrame(df.loc[max_id])
    
    return max_stats
    
```

```python
#used for finding value counts of column variables
def val_counts(column_name):
    
    frequency = df[column_name].value_counts()
    
    
    return frequency
```

```python
def average(column_name):
    
    avg= df[column_name].mean()
    
    return avg
```

```python
def sum(column_name):
    
    sum = df[column_name].sum()
    
    return sum
```

```python
def groupby_sum(column_name1, column_name2):
    
    grp_sum = df.groupby(column_name1)[column_name2].sum()
    
    return grp_sum
```

```python
def groupby_mean(column_name1, column_name2):
    
    grp_mean = df.groupby(column_name1)[column_name2].mean()
    
    return grp_mean
```

```python
def groupby_three(column_name1, column_name2, column_name3):
    
    grp_three = df.groupby(column_name1)[column_name2,column_name3]
    #.sum(), .mean()
    return grp_three
```

```python
def groupby(column_name1, column_name2):
    
    grp = df.groupby(column_name1)[column_name2]
    
    return grp
```

### Research Question 1: What movies generated the most and least profit?
                           
                         a.) What movie made the most and least profit when adjusted for inflation?                          
                         
                         b.) What do adjusted profits for movies look like year over year?

```python
# create new profit column and calculate profit as adjusted rev - adjusted budget

df['profit_adj'] = df['revenue_adj'] - df['budget_adj']
```

```python
#confirm new profit column and calculation
df.head(1)
```

### We created the profits adjusted column to take inflation and real money values into account.  It's a way to make sure that revenues are compared evenly throughout the years.

```python
#lets find out which movies made the most and least profit adjusted for inflation
max('profit_adj')
```

```python
min('profit_adj')
```

### Looks like star wars is the most profitable movie and The Warriors Way is the least profitable movie(both adjusted for inflation).
### I can't say I'm suprised, everyone knows Star Wars, but I've never heard of The Warriors Way.

```python
# profits over the years
import matplotlib.pyplot as plt 

movie_profits = groupby_sum('release_year', 'profit_adj')

plt.plot(movie_profits)
plt.xlabel('Release Year')
plt.ylabel('Combined Yearly Profits in Billions')

plt.title('Movie Profit Trends Per Year', fontsize = 15);
```

### Looks like 2015 had the highest combined profits when adjusted for inflation.  

### Profits have continued on an upward trend since 1960.  
###  The y label values are multiplied by 1e10. 
    2015 profits = 1.35 x 1e10 


```python
movie_profits.idxmax()
```

```python
# Films releases over the years

film_release = groupby('release_year', 'title').count()

plt.plot(film_release)
plt.xlabel('Release Year')
plt.ylabel('Number of Films Release Per Year')
plt.title('Film Release Trends');
```

## RESEARCH QUESTION #3
 ### What Directors have made the most movies.
        3a) Who are the most profitable directors?
        3b) Which directors make most on average per film?
        
        3c) Which movie studio has earned the most revenue?
        3d) How many films do movie studios produce?
        3e) Which studio earns the most on average per film?

```python
#question #1 Directors credited with the most movies
director_count= val_counts('director')

#bar graph characteristics
director_count.iloc[:20].plot(kind='bar',figsize=(10,5),fontsize=15)

#setup the title and the labels 
plt.title('Directors with Most Movies', fontsize = 25)
plt.xticks(rotation = 100, fontsize = 10);
plt.xlabel('Director Name', fontsize = 15)
plt.ylabel('Movie Count', fontsize = 15);
```

### Looks like Woody Allen leads the database with over 40 movies in his name.  

### Other familiar names include Eastwood, Scorcese, and my favorite...Spielberg.

```python
# directors with the highest profits

director_profits = groupby_sum('director', 'profit_adj').sort_values(ascending = False)
#director_profits = dir_profits().sort_values(ascending = False)

director_profits.iloc[:20].plot(kind = 'bar', figsize = (10,5))

plt.title('Directors with Most Profits',fontsize=25)
plt.xticks(rotation=100)
plt.ylabel('Profits',fontsize= 15)
plt.xlabel('Director Name', fontsize = 15);
```

### Speilberg tops the list of highest grossing directors.  I've seen plenty of his movies. I've played a direct part in his profit profile!!!

```python
# 20 Directors with most profit on average per movie

most_reliable = dir_film_avg('director','profit_adj', 'director').sort_values(ascending = False)

most_reliable.iloc[:15].plot(kind = 'barh', figsize = (10,5))

plt.title('Directors who make the most on average per film', fontsize = 25)
plt.xticks(rotation = 100, fontsize = 10);
plt.xlabel('Profit Per Film', fontsize = 15)
plt.ylabel('Director Name', fontsize = 15);
```

### These 15 directors make the most on average per film.  A few of these are also in the top profit profiles for directors as a whole.  Some may classify these 15 as reliable film makers.

```python
# top movie studio revenues
studios = groupby_sum('production_companies', 'profit_adj').sort_values(ascending = False)

studios.iloc[:15].plot(kind = 'barh', figsize = (10,8))

plt.title('Movie Studio Revenues', fontsize = 25)
plt.ylabel('Studio Names')
plt.xlabel('Revenue');
```

### The top 15 movie studios in adjusted revenues.  Paramount is a clear front runner.  Disney is not too far behind.  Considering the sucess of the Marvel Movies under their name, Disney may now be the leader today. 

```python
# Movies Per Studio

studio_films = groupby_sum('title', 'production_companies').value_counts()

studio_films.iloc[:15].plot(kind = 'bar')

plt.title('The Top 15 Film Releasing Studios', fontsize = 25)
plt.xlabel('Film Frequency')
plt.ylabel('Studio Name');
```

### Parmaount again leads the top 15 studios in film production.

```python
# Average Studio Revenue per Film

studio_avg_per_film = dir_film_avg('production_companies', 'profit_adj', 'production_companies').sort_values(ascending = False)[:15]

studio_avg_per_film.plot(kind = 'barh', figsize = (10,8))

plt.title('Average Studio Revenue Per Film', fontsize = 25)
plt.xlabel('Revenue', fontsize = 15)
plt.ylabel('Studio Name', fontsize= 15)
```

### Twentieth Century Fox leads the industry in average revenue per film.
